```python
import gensim
from gensim.models import Word2Vec,KeyedVectors
```

```python
!pip install wget
```

```python
import gensim.downloader as api
print("Downloading word2vec-google-news-300 model through gensim.downloader...")
# This loads the model directly into memory and returns a KeyedVectors object.
# It also handles caching, making it efficient for repeated runs.
# We are storing it in a global variable for the next cell to access.
word_vectors_from_downloader = api.load('word2vec-google-news-300')
print("Model 'word2vec-google-news-300' successfully loaded into memory.")
```

```python
# The model has already been loaded into 'word_vectors_from_downloader' in the previous cell.
# Assign it to 'model' as per the original intent of this cell.
model = word_vectors_from_downloader
print("Model successfully assigned to 'model' variable.")
# The 'limit' parameter is not applicable when directly assigning a pre-loaded model.
# If a specific limit is required for the KeyedVectors, it should be applied to 'word_vectors_from_downloader'
# (e.g., by creating a subset) before this assignment.
```

```python
model['men'].shape
```

```python
model.most_similar('men')
```

```python
model.most_similar('cricket')
```

```python
model.similarity('men','women')
```

```python
model.similar_by_key('india',topn=5)
```

```python
model.similarity('ipl','cricket')
```

```python
model.doesnt_match(['monkey','jungle','book'])
```

```python
model.doesnt_match(['php','python porgramming','java','c'])
```

```python
vec = model['king'] - model['man'] + model['woman']
model.most_similar([vec])
```
